This notebook is for the exploration of the numpy busday

```python
import numpy as np
import pandas as pd
```

```python
#variables
startDate = '2025-12-22'
endDate = '2025-12-31'
weekdays_active = "Mon Tue Wed Thu Fri"
holidays = ['2025-12-24', '2025-12-25', '2025-12-30']
```

```python
#busdaycount example
num_bus_days = np.busday_count(startDate, endDate, weekmask=weekdays_active, holidays=holidays)
print(f"Number of business days between {startDate} and {endDate} is {num_bus_days}")
```

```python
#busday_count but using busdaycalendar
busDays = np.busdaycalendar(weekmask=weekdays_active, holidays=holidays)
cal_bus_days = np.busday_count(startDate, endDate, busdaycal=busDays)
print(f"Number of business days between {startDate} and {endDate} using busdaycalendar is {cal_bus_days}")
```

```python
#generating the date series using pandas
date_range = pd.date_range(start=startDate, end=endDate, freq='D') #this will return an index
date_range = date_range.values.astype('datetime64[D]') #convert to numpy array of datetime64[D]
type(date_range)
```

```python
#determin if the range of dates are business days
busDayArray = np.is_busday(date_range, busdaycal=busDays)
print(busDayArray)
```

```python
#what is the type?
type(busDayArray)
```

```python
#creating a dataframe
df_holidays = pd.DataFrame({'Dates':date_range, "Is a Business Day": busDayArray})
df_holidays
```

```python
#adding another column where the boolean column is converted into 1s and 0s
df_holidays['Is a Business Day_int'] = df_holidays['Is a Business Day'].astype(int)
df_holidays
```
